```python
import numpy as np
from numpy.random import multivariate_normal
import scipy.stats
import matplotlib.pyplot as plt
```

### Генерация двух выборок, с заданным коэффициентом корреляции.

Зададим дисперсии случайных величин X и Y

```python
var_x = 1
var_y = 1
```

Зададим произвольное значение ковариации случайных величин X и Y.

```python
cov_xy = 0.7
```

Т.к. дисперсия случайных величин X и Y равна 1, то коэффициент корреляции так же будет равен 0.7

```python
cov_matrix = [[var_x, cov_xy],
              [cov_xy, var_y]]
```

```python
samples = multivariate_normal((0, 0), cov_matrix, size=1000)
```

```python
plt.scatter(samples[:,0], samples[:,1])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Проверим коэффициент корреляции для двух случайных величин X и Y.

```python
scipy.stats.pearsonr(samples[:,0], samples[:,1]) 
```

### Генерация двух нелинейно зависимых выборок, с возможностью менять силу зависимости. 

```python
n = 1000
x = np.random.uniform(-2, 2, n)
```

Будем использовать кубическую зависимость. Силу зависимости будем менять при помощи среднеквадратического отклонения шума, который мы будем добавлять к этой функциональной зависимости. Шум зададим с помощью стандартного нормального распределения.

```python
def f(x, noise_std):
    y = x**3
    noise = np.random.normal(loc=0., scale=noise_std, size=x.shape[0])
    y += noise
    return y
```

```python
y1 = f(x, 1)

plt.scatter(x, y1)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
y2 = f(x, 2)

plt.scatter(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
y3 = f(x, 3)

plt.scatter(x, y3)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
